# Batting file

```python
batting = pd.read_csv('/Users/Pipita/money_ball/data/batting.csv', sep=',')
```

```python
bat_col = ['playerID','yearID', 'OBP', 'AB', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'BB', 'HBP', 'SH', 'SF']
```

```python
batting_col = pd.DataFrame(batting, columns=bat_col)
```

```python
batting_col.tail()
```

###  Adding OBP column

```python
col_list1 = ['H', 'BB', 'HBP']
```

```python
col_list2 = ['AB', 'BB', 'HBP', 'SF']
```

```python
batting_col['OBP'] = batting_col[col_list1].sum(axis=1) / batting_col[col_list2].sum(axis=1)
```

```python
batting_col.head()
```

### Cleaning the OBP column not (0's, 1's, None vales, NaN)

```python
batting_OBP_cln = batting_col[(batting_col.OBP != 0) & (batting_col.OBP != 1) & (batting_col.OBP != None) & (batting_col.OBP.notnull())]
```

```python
batting_OBP_cln.head()
```
